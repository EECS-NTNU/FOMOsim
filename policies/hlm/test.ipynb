```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

df = pd.read_csv('avg_branching.csv')
```

```python
df = df.sort_values('avg_violations')
df['a'] = df['filename'].str.split('_').str[2]
df['b'] = df['filename'].str.split('_').str[4].str.split('.').str[0]
# df['b'] = df['filename'].str.split('_')[4].split('.')[0]
```

```python
df = df[['avg_violations', 'avg_time', 'a', 'b']]
```

```python
# Assuming 'df' is your DataFrame
# Pivot your DataFrame as needed (if it is not already in the right format)
df1 = df[df['b'].astype(int) < 6]
pivoted_df = df1.pivot(index='b', columns='a', values='avg_violations')

colors = ['#B4D3BA', '#86A86B', '#B57F50', '#4B543B', '#DB6F10', '#00B307', '#FFB674']

# Start plotting
fig, ax = plt.subplots(figsize=(10, 6))  # You can adjust the size as needed

# Plot each column

for i, column in enumerate(pivoted_df.columns):
    color = colors[i % len(colors)]  # Cycle through the color list
    ax.plot(pivoted_df.index, pivoted_df[column], marker='o', color=color, label=f' = {column}')

# Customize the plot to match the uploaded image
ax.set_title('Number of failed events for various combinations of  and ')
ax.set_xlabel('')
ax.set_ylabel('Failed events')
ax.legend(loc='upper left')

# It looks like the original plot uses a specific tick locator for the x-axis
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

# Show the plot
plt.show()
```
